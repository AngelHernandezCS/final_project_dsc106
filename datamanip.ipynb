```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('public/data/scrubbed.csv')
data = data[data['country'] == 'us']
state_abbreviations = {
    'al': 'Alabama', 'ak': 'Alaska', 'az': 'Arizona', 'ar': 'Arkansas', 'ca': 'California',
    'co': 'Colorado', 'ct': 'Connecticut', 'de': 'Delaware', 'fl': 'Florida', 'ga': 'Georgia',
    'hi': 'Hawaii', 'id': 'Idaho', 'il': 'Illinois', 'in': 'Indiana', 'ia': 'Iowa', 'ks': 'Kansas',
    'ky': 'Kentucky', 'la': 'Louisiana', 'me': 'Maine', 'md': 'Maryland', 'ma': 'Massachusetts',
    'mi': 'Michigan', 'mn': 'Minnesota', 'ms': 'Mississippi', 'mo': 'Missouri', 'mt': 'Montana',
    'ne': 'Nebraska', 'nv': 'Nevada', 'nh': 'New Hampshire', 'nj': 'New Jersey', 'nm': 'New Mexico',
    'ny': 'New York', 'nc': 'North Carolina', 'nd': 'North Dakota', 'oh': 'Ohio', 'ok': 'Oklahoma',
    'or': 'Oregon', 'pa': 'Pennsylvania', 'ri': 'Rhode Island', 'sc': 'South Carolina',
    'sd': 'South Dakota', 'tn': 'Tennessee', 'tx': 'Texas', 'ut': 'Utah', 'vt': 'Vermont',
    'va': 'Virginia', 'wa': 'Washington', 'wv': 'West Virginia', 'wi': 'Wisconsin', 'wy': 'Wyoming',
    # Add other US territories if needed
}

data['state'] = data['state'].map(state_abbreviations)
data = data.dropna(subset=['state'])

data = data.reset_index(drop=True)
data.head()
```

```python
locations = data[['latitude', 'longitude ']]
locations.columns = ['latitude', 'longitude']
locations = locations.astype('float')
locations.to_csv('public/data/locations.csv', index=False)
```

```python
data[['shape','state']].groupby('shape', as_index=False).agg('count').sort_values(by='state', ascending=False)[:3]
```

```python
shape_counts = data.groupby(['state', 'shape']).size().reset_index(name='count')
shape_counts = shape_counts.sort_values(['state', 'count'], ascending=[True, False])
top_shapes = shape_counts.groupby('state').head(3).reset_index(drop=True)
top_shapes
```

```python
top_shapes_data = top_shapes.to_csv(index=False)
with open('top_shapes_data.csv', 'w') as f:
    f.write(top_shapes_data)
```
